```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Task 1
uniform_data = np.random.uniform(0, 100, 50)
print(f"Uniform distribution data: {uniform_data}")
```

```python
# Task 2
# Find the minimum, maximum, and range of the data
min_val = np.min(uniform_data)
max_val = np.max(uniform_data)
data_range = max_val - min_val

print(f"Minimum Value: {round(min_val, 2)}")
print(f"Maximum Value: {round(max_val, 2)}")
print(f"Range: {round(data_range, 2)}")
```

```python
# Task 3
# Mean = 17, Std = 0.2, Size = 1000. Count Normal
normal_data = np.random.normal(17, 0.2, 1000)
print("First 10 values from Normal Distribution:\n", normal_data[:10])
```

```python
# Task 4
# Find the minimum, maximum, and range of the data
min_normal = np.min(normal_data)
max_normal = np.max(normal_data)
range_normal = max_normal - min_normal

print(f"Minimum Value: {round(min_normal, 2)}")
print(f"Maximum Value: {round(max_normal, 2)}")
print(f"Range: {round(range_normal, 2)}")
```

```python
# Task 5
# Create a histogram of the data
# bin = 10
plt.figure(figsize=(10, 6))

# Create histogram
A, X, C = plt.hist(normal_data, bins=10, density=True)

# Generate points for the PDF curve
x = np.linspace(min_normal - 0.5, max_normal + 0.5, 100)
pdf = (1 / (0.2 * np.sqrt(2 * np.pi))) * np.exp(-(x - 17)**2 / (2 * 0.2**2))

# Plot PDF
plt.plot(x, pdf, 'r-', lw=2, label='PDF')

plt.title('Normal Distribution')
plt.xlabel('Values')
plt.ylabel('Density')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
