```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Task 1
# Initialise an array containing 20 random values from the interval [0 .. 100]
random_array = np.random.randint(0, 100, 20)
print(random_array)

# Visualise the values of the data points and plot the dataset by using a Line Chart
plt.plot(random_array)
```

```python
# Task 2
Mean = np.mean(random_array)
Median = np.median(random_array)
Range = np.max(random_array) - np.min(random_array)
Standard_Deviation = np.std(random_array)
print("Mean: ", Mean)
print("Median: ", Median)
print("Range: ", Range)
print("Standard Deviation: ", round(Standard_Deviation, 2))
```

```python
# Task 3
# Initialise a pandas Data Frame with the values from the previously initialised array. Print both data structures and observe any differences in the output
df = pd.DataFrame(random_array)
print(df)
print(random_array)
```

```python
# Task 4
# Using 'describe()' method from pandas
print(df.describe())
```

```python
# Task 5
# Calculate the Interquartile range for the dataset
Q1 = np.percentile(random_array, 25)
Q3 = np.percentile(random_array, 75)
IQR = Q3 - Q1
print("Interquartile Range: ", IQR)
```
