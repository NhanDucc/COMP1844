```python
# Define month names
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", 
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Moscow temperatures (Cold Climate)
moscow_temps = np.array([-10, -9, -5, 1, 7, 14, 17, 16, 11, 3, -3, -8])

# Cairo temperatures (Hot Climate)
cairo_temps = np.array([15, 16, 18, 22, 26, 30, 33, 35, 31, 26, 20, 16])

# 2.1 Line Chart
plt.figure(figsize=(10,5))
plt.plot(months, moscow_temps, marker='o', label="Moscow (째C)", linestyle='-')
plt.plot(months, cairo_temps, marker='s', label="Cairo (째C)", linestyle='--')
plt.title("Average Monthly Temperatures: Moscow vs Cairo")
plt.xlabel("Month")
plt.ylabel("Temperature (째C)")
plt.legend()
plt.grid(True)
plt.show()

# 2.2 Bar Chart
bar_width = 0.4
x = np.arange(len(months))

plt.figure(figsize=(10,5))
plt.bar(x - bar_width/2, moscow_temps, width=bar_width, label="Moscow (째C)", color='blue')
plt.bar(x + bar_width/2, cairo_temps, width=bar_width, label="Cairo (째C)", color='orange')
plt.xticks(ticks=x, labels=months)
plt.title("Average Monthly Temperatures: Moscow vs Cairo")
plt.xlabel("Month")
plt.ylabel("Temperature (째C)")
plt.legend()
plt.show()

# 2.3 Box Plot
data = pd.DataFrame({
    "City": ["Moscow"]*12 + ["Cairo"]*12,
    "Temperature (째C)": np.concatenate([moscow_temps, cairo_temps]),
    "Month": months * 2
})

plt.figure(figsize=(8,6))
sns.boxplot(x="City", y="Temperature (째C)", data=data, palette=["blue", "orange"])
plt.title("Temperature Distribution: Moscow vs Cairo")
plt.show()
```
